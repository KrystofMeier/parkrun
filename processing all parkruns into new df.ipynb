```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
from datetime import datetime
%matplotlib inline
```

```python
df = pd.DataFrame(index=range(0,303))
df['Parkrun id'] = np.nan
df['No. of Runners'] = np.nan
df['Winner'] = np.nan
df['Winner M/F'] = np.nan
df['Winner Age'] = np.nan
df['Avg. Time (s)'] = np.nan
df['Median (s)'] = np.nan
df['Sub 20'] = np.nan
df['M'] = np.nan
df['F'] = np.nan
df['Winner Name'] = np.nan
df['New PB'] = np.nan
```

```python
i = 0
while i <= 302:
    parkrun = pd.read_csv('data/clean/final{}.csv'.format(i+1))
    df['Parkrun id'][i] = i+1
    df['No. of Runners'][i] = len(parkrun.index)
    df['Winner'][i] = parkrun['Time'].min()
    if df['Winner'][i] == 'Unkno':
        df['Winner'][i] = 'Unknown'
    df['Winner M/F'][i] = parkrun['Gender'][0]
    df['Winner Age'][i] = parkrun['Age'][0]
    if df['Winner Age'][i] == 'known':
        df['Winner Age'][i] = 'Unknown'
    df['Avg. Time (s)'][i] = parkrun['Time (s)'].mean()
    df['Median (s)'][i] = parkrun['Time (s)'].median() 
    df['Sub 20'][i] = parkrun['Sub 20'].sum()
    df['M'][i] = (parkrun['Gender'] == 'M').sum()
    df['F'][i] = (parkrun['Gender'] == 'F').sum()
    df['Winner Name'][i] = parkrun['Runner'][0]
    df['New PB'][i] = parkrun['New PB'].sum()
    i = i +1
```

```python
df['Sub 20 (%)'] = (df['Sub 20']/df['No. of Runners']*100).round(2)
```

```python
i = 0
df['Winner (s)'] = np.nan
while i < len(df.index):
    df['Winner (s)'][i] = int(df['Winner'][i].split(':')[0])*60+int(df['Winner'][i].split(':')[1])
    i += 1
```

```python
df.to_csv('data/output/keswick_parkrun.csv')
```
