```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import datetime
import re
%matplotlib inline
```

```python
x = 1
while x <= 303:
    
    parkrun = pd.read_csv('data/parkruns/parkrun{}.csv'.format(x))

    parkrun['New PB'] = np.nan
    parkrun['Age'] = np.nan
    parkrun['First Timer'] = np.nan
    parkrun['Runner'] = np.nan
    parkrun['Time (s)'] = np.nan
    parkrun['Sub 20'] = np.nan
    parkrun['M'] = np.nan
    parkrun['F'] = np.nan


    i = 0
    f = 0
    m = 0
    while i < len(parkrun.index):
        parkrun['First Timer'][i] = 'First' in parkrun['Time'][i]
        parkrun['New PB'][i] = 'New' in parkrun['Time'][i]
        parkrun['Time'][i] = parkrun['Time'][i][:5]
        

        parkrun['Gender'][i] = parkrun['Gender'][i].split()[0]
        if parkrun['Gender'][i] == 'Male':
            parkrun['Gender'][i] = 'M'
            parkrun
        elif parkrun['Gender'][i] == 'Female':
            parkrun['Gender'][i] = 'F'            

        if 'parkruns' in parkrun['parkrunner'][i]:
            surname = [re.findall(r'(\w+?)(\d+)', parkrun['parkrunner'][i].split('|')[0])[0] ][0][0] #splitting string on integer
            res = '{} {}'.format(parkrun['parkrunner'][i].split('|')[0].split()[0], surname) #gluing name+surname back together
            parkrun['Runner'][i] = res
        else:
            pass

        parkrun['Age Group'][i] = parkrun['Age Group'][i][:7]
        if 'JW10' in parkrun['Age Group'][i]:
            parkrun['Age'][i] = '10'
            parkrun['Age Group'][i] = 'JW10'
        elif 'JM10' in parkrun['Age Group'][i]:
            parkrun['Age'][i] = '10'
            parkrun['Age Group'][i] = 'JM10'
        else:
            parkrun['Age'][i] = parkrun['Age Group'][i][2:]
      
        if parkrun['Time'][i] == 'Unkno':
            pass
        else:
            parkrun['Time (s)'][i] = int(parkrun['Time'][i].split(':')[0])*60+int(parkrun['Time'][i].split(':')[1])
            parkrun['Sub 20'] = parkrun['Time'] < '20:00'
        
        i += 1
        
    final = parkrun[['Runner', 'Gender', 'Age Group', 'Age', 'Time', 'First Timer', 'New PB', 'Time (s)', 'Sub 20']]
    final.to_csv('data/clean/final{}.csv'.format(x))
    x +=1
```
