```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
sns.set_style('darkgrid')
```

```python
import glob

path = r'data/clean' # use your path
all_files = glob.glob(path + "/*.csv")

li = []

for filename in all_files:
    df = pd.read_csv(filename, index_col=None, header=0)
    li.append(df)

df = pd.concat(li, axis=0, ignore_index=True)
```

```python
i = 0
while i < len(df.index):
    if df['Age'][i] == 'known':
        df['Age'][i] = 'Unknown'
    i += 1
```

```python
df[df['Gender'] == 'M']['Age'].value_counts()
```

```python
df[df['Gender'] == 'F']['Age'].value_counts()
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Age', data=df.sort_values('Age'))
plt.savefig('data/output/age.png')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Age', data=df[df['Gender'] == 'M'].sort_values('Age'))
plt.savefig('data/output/M_age.png')
```

```python
plt.figure(figsize=(12,6))
sns.countplot(x='Age', data=df[df['Gender'] == 'F'].sort_values('Age'))
plt.savefig('data/output/F_age.png')
```
